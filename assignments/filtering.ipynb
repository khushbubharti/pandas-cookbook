```python
import pandas as pd
```

```python
movie = pd.read_csv('../data/movie.csv')
movie.head()
```

```python
c1 = movie['imdb_score']>8
c2 = movie['content_rating']=='PG-13'
c3 = movie['title_year']<2000
c4 = movie['title_year']>2009
movie1=movie[c1&c2&(c3|c4)]
movie1
```
